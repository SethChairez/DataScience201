## Categorical Data

Pandas has a build in data type called Categorical. This helps encode certain columns or parts of your data as being categorical, rather than just an assortment of strings. Categorical has the advantage of being better for memory and for sorting and organizing data. Here is an example:

````python
# Your code here
split_up_part = re.split('[_@.:]',a_string.title())
split_up_part
for item in split_up_part: 
    cleaned_up = [item.replace('Says','').strip() for item in split_up_part]
cleaned_up

    
````

# 8 Remove all other special characters from the data.


````python
df['Tour title' ] = df['Tour title'].apply(lambda x: x.replace('†','')) 
df['Tour title' ] = df['Tour title'].apply(lambda x: x.replace('‡','')) 
df['Tour title' ] = df['Tour title'].apply(lambda x: x.replace('*',''))
df['Tour title' ] = df['Tour title'].apply(lambda x: x.replace('[21][a]',''))
df['Tour title' ] = df['Tour title'].apply(lambda x: x.replace('[4][a]',''))
````

# 9 Separate the Year(s) column into two "Tour Start" and "Tour End"


````python
#do it in function 
#new column called tour start and end
#use.split could try . apply lamnda, or for loop 
print(type(df['Year(s)']))
````

````python
empty_list = []
for item in df['Year(s)']:   
    #item.split('–')
    empty_list.append(item.split('–'))

start_dates = [] 
end_dates = []
for l in empty_list: 
    start_dates.append(l[0])
    if len(l) == 2: 
        end_dates.append(l[1])
    else: 
        end_dates.append(np.nan)

df['Year Start'] =start_dates
df['End Dates']= end_dates
df
````
