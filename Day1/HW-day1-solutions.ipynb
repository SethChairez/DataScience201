### Then using the data frames we sum the quantity of sets purchased.

The age group from 36-50 purchased the most Lego sets with 313, followed by 26-35 year olds.

```python
#5 Create data frames for each of the ages in the following categories: 
# "18 and under", "19 - 25", "26 - 35", "36 - 50", "51 and over". HINT - use masks

# create 5 masks
mask_u18 = (DF['age'] <= 18)
mask_19_25 = (DF['age'] >= 19) & (DF['age'] <=25)
mask_26_35 = (DF['age'] >= 26) & (DF['age'] <=35)
mask_36_50 = (DF['age'] >= 36) & (DF['age'] <=50)
mask_50u = (DF['age'] >= 50)

# create 5 data frames
DF_u18 = DF[mask_u18].copy()
DF_19_25 = DF[mask_19_25].copy()
DF_26_35 = DF[mask_26_35].copy()
DF_36_50 = DF[mask_36_50].copy()
DF_50u = DF[mask_50u].copy()


#6 Which age group has purchased the highest number of Lego sets.

print('under 18:')
print(DF_u18['quantity'].sum())
print('-----')

print('19-25')
print(DF_19_25['quantity'].sum())
print('-----')

print('26-35')
print(DF_26_35['quantity'].sum())
print('-----')

print('36-50')
print(DF_36_50['quantity'].sum())
print('-----')

print('over 50')
print(DF_50u['quantity'].sum())
```

### Which age group spend the most money on Legos?

Again using our data frames above we print out the sum of the money spent. It looks like ages 36-50 also spent the most money.

```python
#7 Which age group has spent the most money on Legos?

# Add a column for money spent.
DF['money_spent'] = DF['us_price']*DF['quantity']

# Then redo the masks
DF_u18 = DF[mask_u18].copy()
DF_19_25 = DF[mask_19_25].copy()
DF_26_35 = DF[mask_26_35].copy()
DF_36_50 = DF[mask_36_50].copy()
DF_50u = DF[mask_50u].copy()

# Then sum the money spent
print('under 18:')
print(DF_u18['money_spent'].sum())
print('-----')

print('19-25')
print(DF_19_25['money_spent'].sum())
print('-----')

print('26-35')
print(DF_26_35['money_spent'].sum())
print('-----')

print('36-50')
print(DF_36_50['money_spent'].sum())
print('-----')

print('over 50')
print(DF_50u['money_spent'].sum())
```

### Whic Lego theme has made the most money?

Here we select two columns: theme and money_spent. Then we take our data frame, for those columns, and group the data by theme. Finally we can sum the columns remaining (money_spent) and the sort the values so we can easily see which theme comes out on top. The Star Wars theme has made the most money for lego in this data.

```python
#8 Which Lego theme has made the most money for Lego?

colummns = ['theme','money_spent']
DF[colummns].groupby('theme').sum().sort_values('money_spent',ascending=False)
```

### Which area-code spent the most money?

First we have to do some work to get the area code out of the phone number column. Then we can do a similar process to above, grouping by area code and then summing money spent. I looks like unknown regions purchased the most, but area code 956 had the highest amount spent for the ones where an area code was listed. There were actually a lot of nans in the phone number columns.

It might actually be more interesting to look at average money spent. Using a for loop we went through the data and calculate the average amount that each but dividing the money_spent by the number of observations for that column. In this case the average spent per customer is highest in area code 212. This is the Manhattan borough of New York City

```python
#9 Which area code has spent the most money on Legos? 
# In the US the area code is the first 3 digits of a phone number.

DF['area_code'] = DF['phone_number'].apply(lambda x: str(x).split('-')[0])

columns = ['area_code','money_spent']
DF[columns].groupby('area_code').sum().sort_values('money_spent',ascending=False)
```

```python
counts = DF['area_code'].value_counts()
counts
```

```python
spent = DF[columns].groupby('area_code').sum().sort_values('money_spent',ascending=False)

average_spent = []
for c in counts.items():
    average_spent.append(spent.loc[c[0]]['money_spent']/c[1])
    # if c[0] != 'nan':
    #     print(f'Area code {c[0]} spent an average of {spent.loc[c[0]]['money_spent']/c[1]}')

spent['average_spent'] = average_spent
spent.sort_values('average_spent',ascending=False)
```
